```python
import ast
```

```python
def convert(text):
    L = []
    for i in ast.literal_eval(text):
        L.append(i['name']) 
    return L 
```

```python
movies['genres'] = movies['genres'].apply(convert)
movies.tail()
```

```python
movies['keywords'] = movies['keywords'].apply(convert)
movies.tail()
```

```python
movies['cast'][0]
# I want the top 3 cast. I need first 3 dictionary 'name'. I need original name not the character name.
```

```python
def convert3(text):
    L = []
    counter = 0
    for i in ast.literal_eval(text):
        if counter < 3:
            L.append(i['name'])
        counter+=1
    return L 
```

```python
movies['cast'] = movies['cast'].apply(convert3)
movies.head()
```

```python
# movies['cast'] = movies['cast'].apply(lambda x:x[0:3])
```

```python
def fetch_director(text):
    L = []
    for i in ast.literal_eval(text):
        if i['job'] == 'Director':
            L.append(i['name'])
    return L 
```

```python
movies['crew'] = movies['crew'].apply(fetch_director)
movies.tail()
```

```python
movies['overview'][0]
```

```python
movies['overview'] = movies['overview'].apply(lambda x:x.split())
movies.tail()
```

```python
# remove spaces from each strings from each list fromeach col from the dataframe movies
```

```python
def collapse(L):
    L1 = []
    for i in L:
        L1.append(i.replace(" ",""))
    return L1
```

```python
movies['cast'] = movies['cast'].apply(collapse)
movies['crew'] = movies['crew'].apply(collapse)
movies['genres'] = movies['genres'].apply(collapse)
movies['keywords'] = movies['keywords'].apply(collapse)
```

```python
movies.tail()
```

```python
movies['tags'] = movies['overview'] + movies['genres'] + movies['keywords'] + movies['cast'] + movies['crew']
movies.tail()
```

```python
new = movies.drop(columns=['overview','genres','keywords','cast','crew'])
new.tail()
```

```python
new['tags'] = new['tags'].apply(lambda x: " ".join(x))
new.tail()
```

```python
new['tags'] = new['tags'].apply(lambda x: x.lower())
new.tail()
```

```python
from sklearn.feature_extraction.text import CountVectorizer
cv = CountVectorizer(max_features=5000,stop_words='english')
# taking only most frequent 5000 words. I can take more or less also. Remove the stopwords of english
```

```python
vector = cv.fit_transform(new['tags']).toarray()
```

```python
vector.shape
```

```python
cv.get_feature_names_out()
# There are some words where Love, Loves, Loved, Loving, etc are there so stemming those
```

```python
import nltk
from nltk.stem.porter import PorterStemmer
ps=PorterStemmer()

def stem(text):
    y=[]
    for i in text.split():
        y.append(ps.stem(i))
    return " ".join(y)

new['tags']=new['tags'].apply(stem)
new.tail()
```

```python
from sklearn.feature_extraction.text import CountVectorizer
cv = CountVectorizer(max_features=5000,stop_words='english')
# taking only most frequent 5000 words. I can take more or less also. Remove the stopwords of english
print(vector.shape)
vector = cv.fit_transform(new['tags']).toarray()
cv.get_feature_names_out()
```

```python
from sklearn.metrics.pairwise import cosine_similarity
# I am not using Euclidean distance. Rather I am using angular distance
# Euclidean distance is not a good measure for high dimension
```

```python
similarity = cosine_similarity(vector)
```

```python
def recommend(movie):
    index = new[new['title'] == movie].index[0]
    distances = sorted(list(enumerate(similarity[index])),reverse=True,key = lambda x: x[1])
    for i in distances[1:6]:
        print(i[0])

# Gives the index of the recomended movies
```

```python
recommend('Gandhi')
```

```python
def recommend(movie):
    index = new[new['title'] == movie].index[0]
    distances = sorted(list(enumerate(similarity[index])),reverse=True,key = lambda x: x[1])
    for i in distances[1:6]:
        print(new.iloc[i[0]].title)

# Gives the name of the recomended movies
```
